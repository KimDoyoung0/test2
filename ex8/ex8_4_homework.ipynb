```python
#파일 불러오기

import csv

with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)

data
```

```python
#숫자로 변환, 결측치 처리
def to_num(value):      #숫자 변환 함수
    try:
        return int(value)   #value값을 int로 반환
    except:
        return None         #value값이 없을시 None으로 반환

for i in data:      #숫자 변환
    i["나이"] = to_num(i["나이"])
    i["국어"] = to_num(i["국어"])
    i["영어"] = to_num(i["영어"])

data
```

```python
#나이 이상치 처리
a_count=0
age=0

#나이 평균 구하기
for i in data:
    if i["나이"] is not None:
        age+=i["나이"]
        a_count+=1
#print(a_count)
av_age=int(age/a_count) 

for i in data:
    if i["나이"] is None:
        i["나이"]=av_age

#data
```

```python
#국어 이상치 처리
g_count=0
G=0

for i in data:
    if i["국어"] is not None:
        G+=i["국어"]
        g_count+=1
av_G=int(G/g_count) 

for i in data:
    if i["국어"] is None:
        i["국어"]=av_G
    if (i["국어"]<0) or (i["국어"]>100):
        i["국어"]=av_G

data
```
