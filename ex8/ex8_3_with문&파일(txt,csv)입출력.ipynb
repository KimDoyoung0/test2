```python
# with문 사용하지 않는 경우
f = open("students.csv", "r", encoding="utf-8")
print(f.read())
# print(10/0)    #이런 오류가 발생하면 밑에 있는 f.close가 실행되지 않는다는 문제가 발생함.
f.close()   # 직접 닫아야 함. 실수로 빼먹거나, 실행전 에러가 발생하면 메모리 누수가 발생한다. 
```

```python
# try-except-finally 문으로 자원 해지
try:
    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
    print(10/0)    
finally:
    f.close()   # 이러면 예외가 발생해도 f.close가 가능
    print("자원 해지 성공")
```

```python
# with문을 사용하는 경우
with open("students.csv", "r", encoding="utf-8") as f:
    print(f.read())
    print(10/0)
```

```python
with open("students.txt", "r", encoding="utf-8") as f:
    # 한 줄씩 읽기.
    for line in f: 
        print(type(line))
        print(line) # 라인 끝의 줄바꿈이 문자 포함되어 두번 줄바꿈이 일어난다(print함수도 줄바꿈이 기본값)
        # 한 라인을 하나의 문자열 가져온다. 라인끝에 줄바꿈 문자까지 포함해서 가져온다
```

```python
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip()) # line에 양쪽 공백제거(줄바꿈포함)하여 print함수의 기본 줄바꿈만으로 출력
```

```python
# 한 라인을 구분자","를 이용하여 두개의 문자열 값으로 저장
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        print(name, int(score))
```

```python
# 딕셔너리 만들기
students = {}

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        students[name] = int(score)

print(students)
```

```python
names = []
scores = []

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        names.append(name)
        scores.append(score)

print(names)
print(scores)
```

```python
import csv

# open()은 내장함수
with open("students.csv", "r", encoding="utf-8") as f:
    # reader(): csv모듈 안에 있는 함수
    reader = csv.reader(f)  
    print(type(reader))

    for row in reader:
        print(row)
        print(type(row))
```

```python
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)  #헤드라인: key, 한줄의 내용: 값

    for row in reader:
        print(row)
```

```python
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)

print(data)     #index로 접근이 가능해짐

data[2]["영어"] = 80

print(data[2])
print(data)
```

```python
import csv

# students_result.csv 파일을 쓰기(w) 모드로 열기
# newline="" : 빈 줄이 추가되는 현상 방지
# encoding="utf-8-sig" : 한글이 엑셀에서 깨지지 않도록 저장

with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())

    # 헤더(컬럼명) 저장
    writer.writeheader()   
    
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")
```
