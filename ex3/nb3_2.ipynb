```python
a = 5
b = 2

print(a + b)   # 7
print(a // b)  # 2
print(a % b)   # 1
print(5 ** 2)
```

```python
print(5 > 3)      # True
print(5 == 3)     # False
```

```python
print(True and False)   # False
print(True or False)    # True
print(not True)         # False

print(10>5 and 10 == 7)     #True and False -> False
```

```python
print(3 in [1, 2, 3])       # True
a = [1, 2, 3]
print(4 in a)  # False
print(4 not in a)   # True
```

```python
a = [1, 2]
b = a

a is b      # True      같은 것을 가리키고 있음. 주소가 같다
a == b      # True      식 == 식. True

c = [1, 2]
d = [1, 2]

c is d      # False     is: 객체 비교
c == d      # True      ==: 객체의 값만 비교
```

```python
score = 50
result = "합격" if score >= 60 else "불합격"    #삼항연산자: result = "a" if (조건식) else "b"
print(result)
```

```python
a = 10

# a = a+ 5와 같은 표현(편한거 사용하면 된다) 
a += 5  # a=a+5
print(a)  # 15
```
