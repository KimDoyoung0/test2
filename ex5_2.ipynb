```python
import numpy as np

a = np.array([[1, 2],
              [3, 4]])
print(a)

b = np.array([[5, 6],
              [7, 8]])
print(a + b) 
# [1+5  2+6] 
# [3+7  4+8]


c = np.array([[5, 6],
              [7, 8]])
d = np.array([[1, 2],
              [3, 4]])
print(c - d)


e = np.array([[1, 2],
              [3, 4]])
print(e * 10)

"""
# 행렬 곱(Matrix Multiplication) 
- 앞 행렬의 열의 수와 뒤 행렬의 행의 수가 같아야
- 결과 행은 앞 행렬의 (행의 수 x 뒤 행렬의 열의 수)를 가지는 행렬이 된다.
- 2행3열 @ 3행1열 --> "2행1열"의 결과 행렬이 반환 된다. 
- 여기서 중요한 건 행렬곱연산은 행렬의 사이즈를 줄일 수 있다는 것이다. 
- 데이터 과학에선 입력값의 특성(피처, 변수)이 너무 많을때 압축할때 행렬곱연산을 사용한다.  
"""

import numpy as np
a = np.array([[1, 2, 3],
              [4, 5, 6]])   # 2행3열

b = np.array([[10],
              [20],
              [30]])        # 3행1열

print(a @ b) # 결과는 2행1열
# [1×10 + 2×20 + 3×30]
# [4×10 + 5×20 + 6×30]

# 전치행렬
a = np.array([[1, 2, 3],
              [4, 5, 6]])

print(a.T)
```
