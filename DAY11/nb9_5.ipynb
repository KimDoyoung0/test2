```python
"""
names=[]
kors=[]
engs=[]

for i in data:
    names.append(i["이름"])
print(names)
"""

names = [i["이름"]for i in data]
kors = [i["국어"]for i in data]
engs = [i["영어"]for i in data]

print(names)
print(kors)
print(engs)
```

8. 국어와 영어의 평균을 계산하여 다음과 같은 자료구조를 생성하시오.
[
('홍길동',87.5),
('이순신',89.0),
...
]

```python
#name_kor_eng = zip(names, kors, engs)
#print(list(name_kor_eng))
name_avg_list = [(n, (int(k)+int(e))/2) for n, k, e in zip(names, kors, engs)]
print(name_avg_list)
```

```python
print(name_kor_list)
print(name_eng_list)
print(name_avg_list)
```
