```python
names = ["홍길동", "이순신", "강감찬"]

for i, name in enumerate(names):
    print(i+1, name)    #(0+1, 홍길동)
    print(f"({i+1}) {name}")
```

```python
nums = [1, 2, 3]
print(list(reversed(nums)))
```

```python
nums = [0, 0, 3]
print(any(nums))

nums = [1, 2, 3]
print(all(nums))

nums = ["", "2", "3"]
print(all(nums))

nums = [{}, {}, {}]
print(any(nums))
```

```python
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
gender = ["F", "M", "F"]
result = zip(names, scores, gender)
print(list(result))
```

```python
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

result = dict(zip(names, scores))

print(result) # {'홍길동': 90, '이순신': 80, '김유신': 100}
```
