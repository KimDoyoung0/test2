```python
import datetime
import time
```

```python
now = datetime.datetime.now()
print(now)
```

```python
print(now.year)
print(now.month)
print(now.day)
print(now.hour)
print(now.minute)
```

```python
today = datetime.date.today()
print(today)
```

```python
print(today.weekday())     #월요일은 0, 화요일은 1, ... 6은 일요일
```

```python
weekday = today.weekday()

week = ["월", "화", "수", "목", "금", "토", "일"]

print(week[weekday])

xmas = datetime.date(2026,12,25)
print(week[xmas.weekday()])
```

```python
print(xmas-today)

print(today+datetime.timedelta(days=1000))

```
