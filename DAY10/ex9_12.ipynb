```python
def outer():
    def inner():
        print("inner")
    return inner
f=outer

f()

outer()()
#outer()는 함수 반환, 그래서 한 번 더 함수 계산
```

```python
def plus(x):
    return x+10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))
```

```python
def add(x, y):
    return x + y

a = lambda x:x+10   #변수에 저장 -> 1급 함수 기능을 포함함

def sub(x, y):
    return x - y

calc = {
    "+": add,
    "-": sub
}

print(calc["+"](10, 5))
print(calc["-"](10, 5))
```
