```python
# 로또 추첨기 만들기

#공을 45개 만들기 == 변수를 45개 만들기
balls = list(range(1,46))
print(balls)
```

```python
#balls shuffle
for i in range(0, 100000):
    temp = balls[0]
    r_nums = int(r.random() * len(balls))  #0.000~44.999 -> 0~44
    balls[0] = balls[r_nums]
    balls[r_nums] = temp

print(balls[:6])
```

```python
# random 모듈 안 sample()함수를 사용
# print(r.sample(range(1,46),6))

lotto = r.sample(range(1,46),6)
print(lotto)

lotto.sort()
print(lotto)
```

```python
for i in range(10):
    lotto = r.sample(range(1,46),6)
    lotto.sort()
    print(f"{i+1}번째 로또 번호: {lotto}")
```

```python
def my_lotto_create(a):
    result = []
    for i in range(a):
        lotto = r.sample(range(1,46),6)
        lotto.sort()
        #print(f"{i+1}번째 로또 번호: {lotto}")
        result.append(tuple(lotto))
    return tuple(result)
```

```python
result = my_lotto_create(3)
print(result)
```

```python
def dice_cast():
    d1 = r.randint(1,6)
    d2 = r.randint(1,6)
    return d1+d2

#person1_dice = {"name": "홍길동", "dice_value":0}
#person1_dice["dice_value"] = dice_cast()
person1_dice = dice_cast()
person2_dice = dice_cast()

if person1_dice == person2_dice:
    print(f"person1:{person1_dice}, person2:{person2_dice} \n비겼습니다.")
elif person1_dice > person2_dice:
    print(f"person1:{person1_dice}, person2:{person2_dice} \nperson1이 이겼습니다.")
else:
    print(f"person1:{person1_dice}, person2:{person2_dice} \nperson2가 이겼습니다.")
```
